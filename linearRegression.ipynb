```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cardata.csv')
```

```python
data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.array(data['Present_Price']).reshape((-1, 1))
x
```

```python
y = np.array(data['Kms_Driven'])
y
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
r_sq
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x, y, color ='b')
plt.plot(x, y_pred, color ='k')
plt.show()
```
